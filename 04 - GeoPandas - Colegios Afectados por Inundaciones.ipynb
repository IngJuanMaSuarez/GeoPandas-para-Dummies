```python
import geopandas as gpd
```

```python
departamentos = gpd.read_file(r"C:\Users\JuanMa\Google Drive\Mis Videos en Youtube\GeoPandas\04 - GeoPandas - Colegios Afectados por Inundaciones\Departamentos.zip")
departamentos.head(5)
```

```python
departamentos.plot()
```

```python
rios = gpd.read_file(r"C:\Users\JuanMa\Google Drive\Mis Videos en Youtube\GeoPandas\04 - GeoPandas - Colegios Afectados por Inundaciones\DrenajeDoble.zip")
rios.head(5)
```

```python
colegios = gpd.read_file(r"C:\Users\JuanMa\Google Drive\Mis Videos en Youtube\GeoPandas\04 - GeoPandas - Colegios Afectados por Inundaciones\EstablecimientosEducativos.zip")
colegios.head(5)
```

```python
antioquia = departamentos.query("DeNombre == 'Antioquia'")
antioquia
```

```python
antioquia.plot()
```

```python
rios_antioquia = gpd.overlay(antioquia, rios, how = "intersection")
rios_antioquia.plot(figsize = (20, 12))
```

```python
axis = antioquia.plot(figsize = (20, 12), alpha = 0.5)
rios_antioquia.plot(ax = axis, color = "red")
```

```python
colegios_antioquia = colegios.query("DeNombre == 'ANTIOQUIA'")
colegios_antioquia.head(5)
```

```python
colegios_antioquia.plot()
```

```python
axis = antioquia.plot(figsize = (20, 12), alpha = 0.5)
colegios_antioquia.plot(ax = axis, color = "white")
rios_antioquia.plot(ax = axis, color = "red")
```

```python
rios_antioquia.crs
```

```python
rios_antioquia["buffer"] = rios_antioquia.buffer(500)
```

```python
rios_antioquia["buffer"].plot(figsize = (20, 12))
```

```python
rios_antioquia = rios_antioquia.set_geometry("buffer")
rios_antioquia.plot(figsize = (20, 12))
```

```python
colegios_afectados = gpd.overlay(colegios_antioquia, rios_antioquia, how = "intersection")
colegios_afectados.head()
```

```python
axis = colegios_antioquia.plot(figsize = (20, 12))
colegios_afectados.plot(ax = axis, color = "red")
rios_antioquia.plot(ax = axis, color = "black")
```
